```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Data Preprocessing
#Import Dataset and build transaction list
dataset = pd.read_csv('TV Shows Data.csv', header = None)
transactions = []
for i in range(0, 9690):
  transactions.append([str(dataset.values[i,j])for j in range(0,29)])
```

```python
#Training the model
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

```python
#Displaying results
results = list(rules)
results
```

```python
#Organizing results to be readable

#Play with indexes to find the TV Shows
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])

resultsinDataFrame
```

```python
#Organizing by Lift Ratio

resultsinDataFrame.nlargest(n = 12, columns = 'Lift')
```
